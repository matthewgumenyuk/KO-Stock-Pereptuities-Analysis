```python
import pandas as pd
import torch
import numpy as np

df = pd.read_excel('COKE_STOCK_DATA.xlsx')
#Date	Close/Last	Volume	Open	High	Low
#STRUCTURE OF EXCELL DATA
#print(df)
#np_array = df.to_numpy(dtype=np.float32)
#print(df[['Close/Last']])
```

```python
#print(df)
std_Array = df.std(axis=0)
avg_Array = df.mean(axis=0)
```

```python
import matplotlib.pyplot as plt
#avg_Array['Close/Last']+std_Array['Close/Last]
def lineWithSTD(dataTarget, degreeDeviation):
#dataTarget = 'Close/Last'
    print(dataTarget +" price Data")
    plt.axhline(y=avg_Array[dataTarget], color='r', linestyle='-', linewidth=1.5)
    plt.axhline(y=avg_Array[dataTarget]+std_Array[dataTarget]*degreeDeviation, color='orange', linestyle='--', linewidth=1.5)
    plt.axhline(y=avg_Array[dataTarget]-std_Array[dataTarget]*degreeDeviation, color='orange', linestyle='--', linewidth=1.5)
    plt.plot(df["Date"], df[dataTarget], linestyle = '-')
    plt.show()
```

```python
deviation_count = 1.5
lineWithSTD('Close/Last',deviation_count)
lineWithSTD('Open',deviation_count)
lineWithSTD('High',deviation_count)
lineWithSTD('Low',deviation_count)
```
